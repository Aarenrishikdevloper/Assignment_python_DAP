```python
import pandas as pd  
london_housing_data = pd.read_csv('London_Housing.csv')   
london_housing_data['date']= pd.to_datetime(london_housing_data["date"])       
london_housing_data['year']= london_housing_data["date"].dt.year      
england_data = london_housing_data[london_housing_data['area'] == 'england']  
max_avg_price_per_year = england_data.groupby('year')['average_price'].max()    
min_avg_price_per_year = england_data.groupby('year')['average_price'].min()   
print("Max average price:")  
print(max_avg_price_per_year)
print("\nMin average price:")     
print(min_avg_price_per_year)

        




```

```python
import pandas as pd  
london_housing_data = pd.read_csv('London_Housing.csv')     
max_crimes_per_area = london_housing_data.groupby('area')['no_of_crimes'].max()   
min_crimes_per_area = london_housing_data.groupby('area')['no_of_crimes'].min()    
print("Max no of crimes per area:")  
print(max_crimes_per_area)
print("\nMin no of crimes per area:")     
print(min_crimes_per_area)        


```

```python
import pandas as pd   
london_housing_data = pd.read_csv('London_Housing.csv')       
low_price_data = london_housing_data[london_housing_data['average_price'] < 100000]  
area_record_count = low_price_data['area'].value_counts()   
print("Total count of records:")  
print(area_record_count)
```
